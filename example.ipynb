```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
# Assuming 'image' is your binary image with a circle in it
# And you have the circle's center (cx, cy) and radius 'r'

# Step 1: Create a mask for the circle
mask = np.zeros((100,100),dtype=np.uint8)  # Create a mask of zeros with the same shape as the image
cv2.circle(mask, (50, 50), 10, (255), thickness=-1)  # Draw the circle on the mask with fill thickness


median = cv2.GaussianBlur(mask,(25,25),sigmaX=2.5)

plt.imshow(median, cmap="gray")
```
